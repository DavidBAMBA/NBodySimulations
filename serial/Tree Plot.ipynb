```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.patches as patches
```

```python
particles = pd.read_csv("data0.txt", sep="\t", header=0, names=["x", "y", "z", "vx", "vy", "vz", "m"])
particles
```

```python
print(np.max(particles["x"]), np.max(particles["y"]))
```

```python
fig, ax = plt.subplots()
ax.scatter(particles["x"], particles["y"], color="k", s=1)
ax.set_xlim(-9, 9)
ax.set_ylim(-9, 9)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
tree = pd.read_csv("tree.txt", sep="\t")
tree
```

```python
deep= len(tree)
xside = tree["xmax"]-tree["xmin"]
yside = tree["ymax"]-tree["ymin"]
```

```python
from random import randint
colors = []

for i in range(np.max(tree["d"])+1):
    colors.append('#%06X' % randint(0, 0xFFFFFF))
```

```python
fig, ax = plt.subplots()
for ii in range(deep):
    square = patches.Rectangle((tree["xmin"][ii], tree["ymin"][ii]), xside[ii], yside[ii], edgecolor=colors[tree["d"][ii]], facecolor='none')
    ax.add_patch(square)
    
ax.set_xlim(-9, 9)
ax.set_ylim(-9, 9)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
